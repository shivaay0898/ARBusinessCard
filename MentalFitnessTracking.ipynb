```python
df = data.copy()
```

```python
#info about data frame
df.info()
```

```python
# transforming non numeric labels into numeric labels
from sklearn.preprocessing import LabelEncoder
l = LabelEncoder()
for i in df.columns:
    if df[i].dtype == 'object':
        df[i]=l.fit_transform(df[i])
```

```python
df.shape
```

```python
X = df.drop('mental_fitness', axis=1)
y = df['mental_fitness']

from sklearn.model_selection import train_test_split  #used to split training data and testing data
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size = .20, random_state=2)
#random_state simply sets seed to the random generator, so that our train test splits are always deterministic. if you dont seed data it would be it is diffrent each time
```

```python
#training(6840,10)
#6840*80/100=5472
#6840*20/100=1368
print("xtrain: ",xtrain.shape)
print("xtest: ", xtest.shape)
print("\nytrain ",ytrain.shape)
print("ytest: " ,ytest.shape)
```

```python
#linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
lr = LinearRegression()
lr.fit(xtrain, ytrain)
ytrain_pred = lr.predict(xtrain)
mse = mean_squared_error(ytrain, ytrain_pred)#Mean Square Error
rmse = (np.sqrt(mean_squared_error(ytrain, ytrain_pred)))#Root Mean Square Error
r2 = r2_score(ytrain, ytrain_pred)
print("The model performance for training set")
print('MSE is {}'.format(mse))
print('MSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")
```

```python
#Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(xtrain, ytrain)
#Model Evaluation for training set
ytrain_pred = rf.predict(xtrain)
mse = mean_squared_error(ytrain, ytrain_pred)
rmse = (np.sqrt(mean_squared_error(ytrain, ytrain_pred)))
r2 = r2_score(ytrain, ytrain_pred)

print("The model performance for training set")
print('MSE is {}'.format(mse))
print('MSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")
```

```python
#RandomForest
ytest_pred = rf.predict(xtest)
mse = mean_squared_error(ytest, ytest_pred)
rmse = (np.sqrt(mean_squared_error(ytest, ytest_pred)))
r2 = r2_score(ytest, ytest_pred)

print("The model performance for test set for Random Forest")
print('MSE is {}'.format(mse))
print('MSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")
#LinearRegression
ytest_pred = lr.predict(xtest)
mse = mean_squared_error(ytest, ytest_pred)
rmse = (np.sqrt(mean_squared_error(ytest, ytest_pred)))
r2 = r2_score(ytest, ytest_pred)

print("The model performance for testing set Linear Regression")
print('MSE is {}'.format(mse))
print('MSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")
```
